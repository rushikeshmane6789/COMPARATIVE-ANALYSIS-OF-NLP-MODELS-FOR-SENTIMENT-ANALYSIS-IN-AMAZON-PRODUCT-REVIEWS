```python
import pandas as pd
import spacy
import string

# Ensure you're working on a copy to prevent SettingWithCopyWarning
df_clean = df_clean.copy()

# Drop NaN values or replace them with a default string
df_clean = df_clean.dropna(subset=['review_body'])

# Alternatively, to replace NaN with a default string: df_clean['lemmatized'].fillna('default_string', inplace=True)

# 1. Convert to lowercase
df_clean['lemmatized'] = df_clean['review_body'].str.lower()

# 2. Remove punctuation
df_clean['lemmatized'] = df_clean['lemmatized'].str.translate(str.maketrans('', '', string.punctuation))

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

# 3. Lemmatize
def lemmatize_text(text):
    doc = nlp(text)
    return ' '.join([token.lemma_ for token in doc])

df_clean['lemmatized'] = df_clean['lemmatized'].apply(lemmatize_text)

# 4. Remove Stop Words
def remove_stopwords(text):
    doc = nlp(text)
    return ' '.join([token.text for token in doc if not token.is_stop])

df_clean['lemmatized'] = df_clean['lemmatized'].apply(remove_stopwords)

# 5. Remove Digits
def remove_digits(text):
    return ''.join([char for char in text if not char.isdigit()])

df_clean['lemmatized'] = df_clean['lemmatized'].apply(remove_digits)

# Display the processed dataframe
print(df_clean)
```

```python
MAX_LEN = 128  # You can adjust this

class SentimentDataset(Dataset):
    def __init__(self, reviews, targets, tokenizer, max_len):
        self.reviews = reviews
        self.targets = targets
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.reviews)

    def __getitem__(self, item):
        review = str(self.reviews[item])
        target = self.targets[item]

        encoding = self.tokenizer.encode_plus(
            review,
            add_special_tokens=True,
            max_length=self.max_len,
            return_token_type_ids=False,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )

        return {
            'review_text': review,
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'targets': torch.tensor(target, dtype=torch.long)
        }
```

```python
# Install the transformers library
!pip install transformers

import pandas as pd
import numpy as np
import re
from nltk.corpus import wordnet
from sklearn.model_selection import train_test_split
from transformers import BertTokenizer, TFBertForSequenceClassification
import tensorflow as tf
from sklearn.metrics import confusion_matrix, accuracy_score, precision_recall_fscore_support
from sklearn.utils import resample
import random

# Ensure you have the WordNet data downloaded
import nltk
nltk.download('wordnet')

# Text Cleaning
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^a-z0-9\s]', '', text)  # Remove special characters
    return text

df_clean['review_body'] = df_clean['review_body'].apply(clean_text)

# Data Augmentation using Synonym Replacement
def synonym_replace(text):
    words = text.split()
    new_words = words.copy()
    for i in range(len(words)):
        synonyms = wordnet.synsets(words[i])
        if synonyms and len(synonyms) > 0 and synonyms[0].lemmas():
            synonym_words = [lemma.name() for lemma in synonyms[0].lemmas()]
            synonym_choice = random.choice(synonym_words)
            new_words[i] = synonym_choice
    return ' '.join(new_words)

df_augmented = df_clean.copy()
df_augmented['review_body'] = df_augmented['review_body'].apply(synonym_replace)

# Combine original and augmented data
df_combined = pd.concat([df_clean, df_augmented])

# Adjust the labels to start from 0 instead of 1
def adjust_labels(rating):
    if rating in [1, 2]:
        return 0  # Negative
    elif rating == 3:
        return 1  # Neutral
    else:
        return 2  # Positive

df_combined['star_rating'] = df_combined['star_rating'].apply(adjust_labels)

# Split the data into training, validation, and test sets
train_df, temp_df = train_test_split(df_combined, test_size=0.3, random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

# Initialize the model and tokenizer
model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=3)  # Assuming 3 labels: negative, neutral, positive
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Tokenize and prepare the datasets
train_encodings = tokenizer(train_df['review_body'].tolist(), truncation=True, padding=True, max_length=128, return_tensors="tf")
train_labels = tf.convert_to_tensor(train_df['star_rating'].tolist())

val_encodings = tokenizer(val_df['review_body'].tolist(), truncation=True, padding=True, max_length=128, return_tensors="tf")
val_labels = tf.convert_to_tensor(val_df['star_rating'].tolist())

test_encodings = tokenizer(test_df['review_body'].tolist(), truncation=True, padding=True, max_length=128, return_tensors="tf")
test_labels = tf.convert_to_tensor(test_df['star_rating'].tolist())

# Convert encodings to tf.data.Dataset
train_dataset = tf.data.Dataset.from_tensor_slices((dict(train_encodings), train_labels)).shuffle(1000).batch(16)
val_dataset = tf.data.Dataset.from_tensor_slices((dict(val_encodings), val_labels)).batch(16)
test_dataset = tf.data.Dataset.from_tensor_slices((dict(test_encodings), test_labels)).batch(16)

# Define a custom loss function
loss_object = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)

def compute_loss(labels, logits):
    return loss_object(labels, logits)

# Train the model
optimizer = tf.keras.optimizers.Adam(learning_rate=5e-5)
model.compile(optimizer=optimizer, loss=compute_loss, metrics=['accuracy'])
model.fit(train_dataset, validation_data=val_dataset, epochs=3)

# Predict on validation and test sets
val_predictions = np.argmax(model.predict(val_dataset).logits, axis=-1)
test_predictions = np.argmax(model.predict(test_dataset).logits, axis=-1)

# Compute metrics
val_confusion = confusion_matrix(val_df['star_rating'].tolist(), val_predictions)
test_confusion = confusion_matrix(test_df['star_rating'].tolist(), test_predictions)

val_accuracy = accuracy_score(val_df['star_rating'].tolist(), val_predictions)
test_accuracy = accuracy_score(test_df['star_rating'].tolist(), test_predictions)

val_precision, val_recall, val_f1, _ = precision_recall_fscore_support(val_df['star_rating'].tolist(), val_predictions, average='weighted')
test_precision, test_recall, test_f1, _ = precision_recall_fscore_support(test_df['star_rating'].tolist(), test_predictions, average='weighted')

# Print results
print("Validation Confusion Matrix:\n", val_confusion)
print("\nTest Confusion Matrix:\n", test_confusion)

print("\nValidation Metrics:")
print("Accuracy:", val_accuracy)
print("Precision:", val_precision)
print("Recall:", val_recall)
print("F1 Score:", val_f1)

print("\nTest Metrics:")
print("Accuracy:", test_accuracy)
print("Precision:", test_precision)
print("Recall:", test_recall)
print("F1 Score:", test_f1)

```
